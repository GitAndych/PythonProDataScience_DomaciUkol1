```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import OLSInfluence

data = pd.read_csv("df.csv")
data_corr = data.drop(columns=["Company", "Sector"])
```

```python
data_corr.corr() # Největší vliv má ukazatel FL (0.759323)
```

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.ols(formula=formula, data=data_corr)
res = mod.fit()
res.summary()
```

```python
formula = "Y ~ FL + OPM + TAT" # Odebrala jsem statisticky nevýznamné ukazatele CR, DA, RT, DE, PT a nechala jen statisticky významné
mod = smf.ols(formula=formula, data=data_corr)
res = mod.fit()
res.summary()
```

```python
data["Odhad_Y"] = res.predict()
data["Rozdíl"] = data["Odhad_Y"] - data["Y"]

max_hodnota = data["Rozdíl"].idxmax()

max_cely_radek = data.iloc[max_hodnota]
max_cely_radek # Akcie s největším rozdílem odhadu koeficientu a jeho skutečnou hodnotou je z firmy pod názvem SYXLy, 79 řádek
```
