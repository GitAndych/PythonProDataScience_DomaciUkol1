```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import OLSInfluence

data = pd.read_csv("df.csv")
data_corr = data.drop(columns=["Company", "Sector"])
```

```python
# Robustní regrese
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.rlm(formula=formula, data=data_corr)
res = mod.fit()
res.summary()
```

```python
formula = "Y ~ CR + FL + OPM + TAT" # Podle p-hodnoty z tabulky výše jsem vyřadila statisticky nevýznamné koeficienty
mod = smf.rlm(formula=formula, data=data_corr)
res = mod.fit()
res.summary()
```

```python
data["Odhad_Y"] = res.predict()
data["Rozdíl"] = data["Odhad_Y"] - data["Y"]

max_hodnota = data["Rozdíl"].idxmax()

max_cely_radek = data.iloc[max_hodnota]
max_cely_radek # Potenciálně nejvíce podhodnocená akcie je v případě robustní regrese stejná jako v předchozím případě (na 79 řádku)
```
